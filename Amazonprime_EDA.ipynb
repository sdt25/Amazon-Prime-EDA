## 3. ***Data Wrangling***

### Data Wrangling Code

````python
print(amazonprime_df.columns)  # List all column names
````

````python
# Write your code to make your dataset analysis ready.
#  1. Handling Missing Values
amazonprime_df.fillna({'age_certification': 'Unknown',
                  'genres': 'Unknown',
                  'production_countries': 'Unknown'}, inplace=True)  # Fill missing categorical data
amazonprime_df.fillna({'imdb_score': amazonprime_df['imdb_score'].median(),
                  'tmdb_score': amazonprime_df['tmdb_score'].median()}, inplace=True)  # Fill missing numerical data
amazonprime_df['seasons'].fillna(0, inplace=True)  # If missing, set TV show seasons to 0
````

````python
#  2. Removing Duplicates
amazonprime_df.drop_duplicates(subset=['id'], keep='first', inplace=True)
````

````python
#  3. Fixing Data Types
amazonprime_df['release_year'] = pd.to_datetime(amazonprime_df['release_year'],
format='%Y', errors='coerce')  # Convert to datetime
if 'runtime' in amazonprime_df.columns:
  amazonprime_df['runtime'] = amazonprime_df['runtime'].astype(int)  # Convert runtime to integer
````

````python
#  4. Handling Outliers (Capping extreme values)
if 'runtime' in amazonprime_df.columns:
    amazonprime_df['runtime'] = np.clip(amazonprime_df['runtime'], amazonprime_df['runtime'].quantile(0.01), amazonprime_df['runtime'].quantile(0.99))
````

````python
#  5. Standardizing Text Data
amazonprime_df['title'] = amazonprime_df['title'].str.lower().str.strip()
amazonprime_df['genres'] = amazonprime_df['genres'].str.lower().str.strip()
amazonprime_df['production_countries'] = amazonprime_df['production_countries'].str.lower().str.strip()
````

````python
#  6. Feature Engineering
amazonprime_df['decade'] = (amazonprime_df['release_year'].dt.year // 10) * 10  # Create decade column
amazonprime_df['imdb_category'] = pd.cut(amazonprime_df['imdb_score'], bins=[0, 4, 6, 8, 10], labels=['Poor', 'Average', 'Good', 'Excellent'])  # Categorize IMDb scores
````

````python
#  7. Final Check
print("Final Dataset Shape:", amazonprime_df.shape)
print("Missing Values After Cleaning:\n", amazonprime_df.isnull().sum())
````
